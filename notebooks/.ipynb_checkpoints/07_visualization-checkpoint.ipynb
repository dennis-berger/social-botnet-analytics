```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Load data
repo_root    = Path().resolve().parent
nodes = pd.read_csv(repo_root / "neo4j_import/nodes.csv")
edges = pd.read_csv(repo_root / "neo4j_import/edges.csv")
clusters = pd.read_csv(repo_root / "neo4j_import/clusters.csv")

# Merge clusters into nodes and edges
nodes = nodes.merge(clusters, on="user_id")
edges = edges.merge(clusters.rename(columns={"user_id": "source"}), on="source")
edges = edges.merge(clusters.rename(columns={"user_id": "target"}), on="target", suffixes=("_src", "_tgt"))
```

```python
# Pie Chart: Bot vs. Human Distribution
bot_counts = nodes["is_bot"].value_counts().sort_index()
plt.figure()
bot_counts.plot(kind="pie", labels=["Human", "Bot"], autopct="%1.1f%%")
plt.title("Bot vs. Human Distribution")
plt.ylabel("")  # hide y-label
plt.show()
```

```python
# Top 5 Bot-Heavy Clusters (horizontal bar chart)
cluster_stats = nodes.groupby("cluster").agg(
    botRatio=("is_bot", "mean"),
    clusterSize=("user_id", "size")
).reset_index()
top5 = cluster_stats[cluster_stats["clusterSize"] >= 20].sort_values("botRatio", ascending=False).head(5)

plt.figure()
plt.barh(top5["cluster"].astype(str), top5["botRatio"])
plt.xlabel("Bot Ratio")
plt.ylabel("Cluster ID")
plt.title("Top 5 Bot-Heavy Clusters")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Community Cohesion Ratios (line chart)
cohesion = edges.groupby("cluster_src").apply(
    lambda g: (g["cluster_src"] == g["cluster_tgt"]).sum() / len(g)
).reset_index(name="cohesionRatio")
cohesion = cohesion.sort_values("cohesionRatio", ascending=False)

plt.figure()
plt.plot(cohesion["cluster_src"].astype(str), cohesion["cohesionRatio"], marker="o")
plt.xticks(rotation=90)
plt.xlabel("Cluster ID")
plt.ylabel("Cohesion Ratio")
plt.title("Community Cohesion Ratios")
plt.show()
```

```python
# Bridge Nodes by Cluster (scatter plot)
inter = edges[edges["cluster_src"] != edges["cluster_tgt"]].groupby("source").size().reset_index(name="interDeg")
inter = inter.merge(clusters.rename(columns={"user_id": "source"}), on="source")

plt.figure()
plt.scatter(inter["cluster"], inter["interDeg"])
plt.xlabel("Cluster ID")
plt.ylabel("Inter-Cluster Out-Degree")
plt.title("Bridge Nodes by Cluster")
plt.show()
```

```python
# LT Spread per Node (bar chart)
lt = pd.DataFrame({
    "cluster": [39, 38, 65],
    "size": [45, 407, 480],
    "outerInfluence": [460, 3907, 2854]
})
lt["spreadPerNode"] = lt["outerInfluence"] / lt["size"]

plt.figure()
plt.bar(lt["cluster"].astype(str), lt["spreadPerNode"])
plt.xlabel("Cluster ID")
plt.ylabel("Spread per Node")
plt.title("LT Spread per Node")
plt.show()
```

```python
# Classifier Performance (grouped bar chart)
clf = pd.DataFrame({
    "Model": ["Decision Tree", "Naive Bayes", "k-NN (k=5)"],
    "Precision": [0.7927, 0.4707, 0.7515],
    "Recall":    [0.7927, 0.9636, 0.6709],
    "F1-Score":  [0.7927, 0.6325, 0.7089]
})
clf = clf.set_index("Model")

plt.figure()
clf.plot(kind="bar")
plt.ylabel("Score")
plt.title("Classifier Performance on Bot Detection")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
